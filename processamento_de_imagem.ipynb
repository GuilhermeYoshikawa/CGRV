## Processamento de imagem 2 - Matrizes de cores

```python
# Canais
print('Qtde de canais: ', end='')
print(imagem.shape[2])
```

```python
# Importação das bibliotecas
import cv2
```

```python
# Leitura da imagem com a função imread()
imagem = cv2.imread('entrada.png')
```

```python
# Veja que a ordem BGR e não RGB
(b, g, r) = imagem[50, 100]

print('O pixel (50, 100) tem as seguintes cores: ')
print('R - Vermelho: ', r, '\nG - Verde: ', g, '\nB - Azul', b)
```

```python
for y in range(0, imagem.shape[0]): # depois varre todos os valores da vertical
  for x in range(0, imagem.shape[1]): # primeiro varre todos os valores da horizontal
    imagem[y, x] = (255,0,0) # substitui os valores e como a ordem é BGR a cor será azul
```

```python
cv2_imshow(imagem)
```

```python
for y in range(0, imagem.shape[0]): # percorre linhas
  for x in range(0, imagem.shape[1]): # percorre colunas
    imagem[y, x] = (x%256,y%256,x%256) # a porcentagem faz com que pegue o resto do BGR
```

```python
cv2_imshow(imagem)
```

```python
for y in range(0, imagem.shape[0]): # percorre linhas
  for x in range(0, imagem.shape[1]): # percorre colunas
    imagem[y, x] = (0,(x*y)%256,0) # BGR, onde B=0, G=muliplica colunas e linhas e divide por 256, R=0
```

```python
cv2_imshow(imagem)
```

```python
for y in range(0, imagem.shape[0]): # percorre linhas
  for x in range(0, imagem.shape[1]): # percorre colunas
    imagem[y:y+5, x: x+5] = (0,255,255) # BGR com resultado amarelo
```

```python
cv2_imshow(imagem)
```

## Processamento de imagem 3 - Fatiamento

```python
# Importação das bibliotecas
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Leitura da imagem com a função imread()
imagem = cv2.imread('entrada_2.jpg')
```

```python
# Cria um retangulo azul por toda a largura da imagem
imagem[30:50, :] = (255, 0 , 0)
```

```python
# Cria um retangulo amarelo por toda a altura da imagem
imagem[:, 200:220] = (0, 255, 255)
```

```python
# Cria um retangulo verde da linha 150 a 300 nas colunas 250 a 350
imagem[150:300, 250:350] = (0, 255, 0)
```

```python
# Cria um quadrado ciano da linha 150 a 300 nas colunas 250 a 350
imagem[300:400, 50:150] = (255, 255, 255)
```

```python
# Cria um quadrado preto
imagem[70:100, 300: 450] = (0, 0, 0)
```

```python
# importação das bibliotecas

import cv2
from google.colab.patches import cv2_imshow

import numpy as np
```

```python
# Leitura da imagem com a função imread()
imagem = cv2.imread('entrada_2.jpg')
```

```python
# Variável contendo os códigos das cores
vermelho = (0, 0, 255)
verde = (0, 255, 0)
azul = (255, 0, 0)
```

```python
# Cria uma linha verde e vermelho
cv2.line(imagem, (0, 0), (100, 200), verde)
cv2.line(imagem, (300, 200), (150, 150), vermelho, 5)
```

```python
# Cria um retangulo azul e verde
cv2.rectangle(imagem, (20, 20), (120, 120), azul, 10)
cv2.rectangle(imagem, (200, 50), (225, 125), verde, -1)
```

```python
# Cria vários circulos vermelhos com variações do raio
(X, Y) = (imagem.shape[1] // 2, imagem.shape[0] // 2) # pegando o centro perfeito da figura
for raio in range(0, 175, 15):
  cv2.circle(imagem, (X, Y), raio, vermelho)
```
