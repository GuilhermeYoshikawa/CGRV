## PROJETO DE DESEMPENHO

Em python desenvolva o programa de uma agenda simples (nome e telefone) com um menu de opções para a agenda contendo: Incluir um novo contato; Consultar um contato individual; Alterar um contato; Excluir um contato; Mostrar a agenda completa; e sair dela.

```python
from time import sleep

def leiaInt(msg):
  while True:
    try:
      n = int(input(msg))
    except (ValueError, TypeError):
      print('ERRO! Digite uma opção válida.')
      continue
    except (KeyboardInterrupt):
      print('Usuário preferiu não digitar esse número.')
      return 0
    else:
      return n

def linha(tam = 42):
  return '-' * tam

def cabecalho(txt):
  print(linha())
  print(txt.center(42))
  print(linha())

def menu(lista):
  cabecalho('MENU')
  c = 1
  for item in lista:
    print(f'{c} - {item}')
    c += 1
  print(linha())
  opc = leiaInt('Sua opção: ')
  print(linha())
  return opc
```

```python
contatos = []

while True:
  opcao = menu(["Adicionar contato", "Consultar contato", "Alterar contato", "Excluir contato", "Mostrar contatos", "Fechar"])
  
  if opcao == 1:
    cabecalho('ADICIONAR CONTATO')
    nome = str(input("Nome: "))
    telefone = int(input("Telefone: "))
    contatos.extend((nome, telefone))
    print("Contato adicionado!")

  elif opcao == 2:
    cabecalho('CONSULTAR CONTATO')
    consultar = str(input("Nome: "))
    if consultar in contatos:
      print("O contato %s está na lista" %consultar)
    else:
      print("O contato %s não está na lista" %consultar)

  elif opcao == 3:
    cabecalho('ALTERAR CONTATO')
    alterar_contato = str(input("Nome: "))
    if consultar in contatos:
      alterar = str(input("Novo contato: "))
      contatos[contatos.index(consultar)] = alterar
      print("Contato alterado com sucesso!")
    else:
      print("Contato não encontrado na agenda.")

  elif opcao == 4:
    cabecalho('DELETAR CONTATO')
    remover = str(input("Nome: "))
    if consultar in contatos:
      contatos.remove(remover)
      print("Contato deletado.")
    else:
      print("Contato não foi encontrado.")

  elif opcao == 5:
    cabecalho('AGENDA')
    print(contatos)

  elif opcao == 6:
    cabecalho('Saindo do sistema...')
    break
  else:
    print('ERRO! Digite uma opção válida.')
  sleep(1)
```
